# Manga frame detect with OpenCV

In this tutorial, we will use two features of OpenCV: Contours Finding and Convex Hull to recognize the frames of a manga page.

First, we need to import the neccessary libraries:

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

Define a new variable to hold the file name of the manga page:

```python
filename = 'manga.jpg'
```

## Open the manga page

We use `cv2.imread()` function to open the manga page for processing. It's neccessary to convert it to a grayscale image with `cv2.COLOR_BGR2GRAY` parameter for processing.

```python
img = cv2.imread(filename)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

We also create 2 more images for comparing and processing result:

```python
original = cv2.imread(filename)
result = cv2.imread(filename)
```

 Let's see how the image look like:

```python
plt.subplot(111)
plt.imshow(img, cmap='Greys_r')
plt.show()
```

## Thresholding

The first processing step is to thresholding the manga page. This method is also good for removing unecessary content such as watermark, etc... on the manga page, so we can find the frame more accurately.

We also need to use the parameter `cv2.THRESH_BINARY_INV` to invert the color of the manga page after thresholding. It's a must for finding contours.

```python
ret, thresh = cv2.threshold(img, 80, 255, cv2.THRESH_BINARY_INV)
```

Let's print out the thresholded image and compare with the original image:

```python
plt.subplot(121),plt.imshow(original),plt.title('ORIGINAL')
plt.subplot(122),plt.imshow(thresh, cmap='Greys_r'),plt.title('THRESHOLD')
plt.show()
```

As you see, the watermarks are gone! Now, let's find the contours.

## Contours

To find the contours, we use `cv2.findContours()` function:

```python
contours, h = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

Let's see how the contours data look like:

```python
for cnt in contours:
    cv2.drawContours(result, [cnt], -1, 255, 3)
```

```python
plt.subplot(121),plt.imshow(original),plt.title('ORIGINAL')
plt.subplot(122),plt.imshow(result),plt.title('THRESHOLD')
plt.show()
```

It may look good, but just the contours itself doesn't help us find the manga frame, what we can do here is create a convex hull for the contours.

## Convex Hull

We modify the `for..in` loop to create a convex hull for each contours:

```python
for cnt in contours:
    hull = cv2.convexHull(cnt)
    cv2.drawContours(result, [hull], -1, 255, -1)
```

Now let's see how it change:

```python
plt.subplot(121),plt.imshow(original),plt.title('ORIGINAL')
plt.subplot(122),plt.imshow(result),plt.title('THRESHOLD')
plt.show()
```

This look better now, so we have the convex hull surrounding each frames of the manga page.
